```python
import tabula
import numpy as np
import pandas as pd
```

```python
df14 = tabula.read_pdf("BP.pdf", pages='14', pandas_options={'header': [0,1,2]})
df15 = tabula.read_pdf("BP.pdf", pages='15', pandas_options={'header': [0,1,2]})
df16 = tabula.read_pdf("BP.pdf", pages='16', pandas_options={'header': [0,1,2]})
df17 = tabula.read_pdf("BP.pdf", pages='17', pandas_options={'header': [0,1,2]})
df18 = tabula.read_pdf("BP.pdf", pages='18', pandas_options={'header': [0,1,2]})
df19 = tabula.read_pdf("BP.pdf", pages='19', pandas_options={'header': [0,1,2]})
```

```python
df14
```

```python
#tabula.convert_into("BP.pdf", "BP.csv", output_format="csv", pages='17',pandas_options={'header': [0,1,2]})
```

```python
def remove_y(number):
    if number[0] == 'y':
        number = '-' + number[1:]
    return float(number)

def cleaning(df):
    df_copy = df.copy()
    for i in range(4,8):
        df_copy[df_copy.columns[i]] = df_copy[df_copy.columns[i]].apply(remove_y)
    return df_copy
```

```python
df14_cleaned = cleaning(df14)
df15_cleaned = cleaning(df15)
df16_cleaned = cleaning(df16)
df17_cleaned = cleaning(df17)
df18_cleaned = cleaning(df18)
df19_cleaned = cleaning(df19)
```

```python
final_df = pd.concat([df14_cleaned, df15_cleaned, df16_cleaned, df17_cleaned, df18_cleaned, df19_cleaned])
```

```python
col_names = ['No. of compound', 'T_publ (K)', 'T_pred M-235 (K)', 'T_pred M-20 (K)', 'Absolute Error M-235 (K)', 'Absolute Error M-20 (K)','Relative Error M-235 (%)', 'Relative Error M-20 (%)']
final_df.columns = col_names
```

```python
final_df
```

```python
final_df.to_csv('BP1.csv', index=False)
```

```python
df20 = tabula.read_pdf("BP.pdf", pages='20', pandas_options={'header': [0,1]})
df20
```

```python
def first(string):
    string1 = string.split()
    return string1[0]

def second(string):
    string1 = string.split()
    return string1[1]
```

```python
df_cleaning1 = df20.copy()
```

```python
df_cleaning1.insert(0, 'Name', df20[df20.columns[0]].apply(second))
```

```python
df_cleaning1
```

```python
df_cleaning1.insert(0, 'No.', df20[df20.columns[0]].apply(first))
```

```python
df_cleaning1
```

```python
df_cleaning2 = df_cleaning1.drop(columns=df_cleaning1.columns[2])
```

```python
df_cleaning3 = df_cleaning2.drop(columns=df_cleaning1.columns[4])
```

```python
df_cleaning3
```

```python
def max_three_characters(string):
    if len(string) > 3:
        string = string[1:]
    return string
```

```python
df_cleaning3['No.'] = df_cleaning3['No.'].apply(max_three_characters)
```

```python
def remove_y_with_na(number):
    if type(number) == str:
        if number[0] == 'y':
            number = '-' + number[1:]
        return float(number)
    else:
        return number
    
def cleaning_with_na(df, start_index, num_col):
    df_copy = df.copy()
    for i in range(6,10):
        df_copy[df_copy.columns[i]] = df_copy[df_copy.columns[i]].apply(remove_y_with_na)
    return df_copy
```

```python
df_cleaning4 = cleaning_with_na(df_cleaning3)
```

```python
col_names = ['No.', 'Name', 'T_calc (K)', 'T_publ (K)', 'T_pred M-235 (K)', 'T_pred M-20 (K)', 'Absolute Error M-235 (K)', 'Absolute Error M-20 (K)','Relative Error M-235 (%)', 'Relative Error M-20 (%)']
df_cleaning4.columns = col_names
```

```python
df_cleaning4
```

```python
df_cleaning4.to_csv('BP2.csv', index=False)
```
